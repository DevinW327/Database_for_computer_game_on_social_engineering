```python
#function to connect to server
def create_server_connection(host_name, user_name, user_password):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
```

```python
#function to create a database
def create_database(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("Database created successfully")
    except Error as err:
        print(f"Error: '{err}'")
```

```python
#function to connect to database
def create_db_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
```

```python
#function to execute query
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query successful")
    except Error as err:
        print(f"Error: '{err}'")
```

```python
#function to read results
def read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as err:
        print(f"Error: '{err}'")
```

```python
import mysql.connector
from tkinter import Tk, Button, Label, messagebox, simpledialog, StringVar, Frame, Entry

cursor = connection.cursor(dictionary=True)

# Utility Functions
def fetch_questions():
    #Fetch all questions from the Quiz table in random order
    cursor.execute("SELECT * FROM Quiz ORDER BY RAND()")
    return cursor.fetchall()

def update_user_stats(user_id, score):
    #Update user statistics after a game session
    cursor.execute("SELECT high_score, avg_score, num_times_played FROM Users WHERE user_id = %s", (user_id,))
    user = cursor.fetchone()

    if user:
        high_score = max(user['high_score'], score)
        num_times_played = user['num_times_played'] + 1
        avg_score = ((user['avg_score'] * user['num_times_played']) + score) / num_times_played
        
        cursor.execute("""
            UPDATE Users
            SET high_score = %s, avg_score = %s, last_score = %s, num_times_played = %s
            WHERE user_id = %s
        """, (high_score, avg_score, score, num_times_played, user_id))
        connection.commit()

# Tkinter Application
class GameApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Social Engineering Quiz Game")

        self.window_size = "800x600"  # Small size
        self.root.geometry(self.window_size)
        
        self.bg_color = "#f0f0f0"  # Default background color (light mode)
        self.button_color = "#ffffff"  # Default button color
        self.text_color = "#000000"  # Default text color
        
        self.current_user_id = None
        self.questions = []
        self.current_question_index = 0
        self.score = 0
        
        self.question_var = StringVar()
        self.answer_buttons = []

        self.font_size_header = 18  # Font size for headers (larger)
        self.font_size_label = 14   # Font size for labels (medium)
        self.font_size_button = 12  # Font size for buttons (smaller)

        self.setup_main_menu()

    def setup_main_menu(self):
        #Setup the main menu with buttons for login, account creation, and guest play
        self.clear_window()
        menu_frame = Frame(self.root, padx=20, pady=20, bg=self.bg_color)
        menu_frame.pack(pady=50)

        Label(menu_frame, text="Welcome to the Social Engineering Quiz Game", font=("Helvetica", self.font_size_header, "bold"), bg=self.bg_color, fg=self.text_color).pack(pady=10)
        
        Button(self.root, text="Login", font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color, command=self.login).pack(pady=5)
        Button(self.root, text="Create Account", font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color, command=self.create_account).pack(pady=5)
        Button(self.root, text="Play as Guest", font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color, command=self.play_as_guest).pack(pady=5)
        Button(self.root, text="Leaderboard", font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color, command=self.leaderboard).pack(pady=5)
        Button(self.root, text="Exit", font=("Helvetica", self.font_size_button), command=self.ask_stop, bg=self.button_color, fg=self.text_color).pack(pady=5)

        Button(menu_frame, text="Toggle Dark/Light Mode", font=("Helvetica", self.font_size_label), bg=self.button_color, fg=self.text_color, 
           command=self.toggle_theme).pack(pady=5)
        Button(menu_frame, text="Toggle Window Size", font=("Helvetica", self.font_size_label), bg=self.button_color, fg=self.text_color, command=self.toggle_window_size).pack(pady=5)
        Button(menu_frame, text="Change Text Size", font=("Helvetica", self.font_size_label), bg=self.button_color, fg=self.text_color, command=self.change_text_size).pack(pady=5)

    def clear_window(self):
        #Clear all widgets from the window
        for widget in self.root.winfo_children():
            widget.destroy()

    def toggle_theme(self):
        #checks which mode the game is and switches to the opposite
        if self.bg_color == "#f0f0f0":
            self.bg_color = "#333333"
            self.button_color = "#000000"
            self.text_color = "#ffffff"
        else:
            self.bg_color = "#f0f0f0"
            self.button_color = "#ffffff"
            self.text_color = "#000000"
            
        self.root.config(bg=self.bg_color)
        self.setup_main_menu()

    def toggle_window_size(self):
        # Toggle the window size between "800x600" (small) and "1200x900" (large)
        if self.window_size == "800x600":
            self.window_size = "1200x900"  # Larger size
        else:
            self.window_size = "800x600"  # Smaller size
            if self.font_size_button>14:
                self.font_size_header =18
                self.font_size_label =16
                self.font_size_button = 14
                self.setup_main_menu()

        self.root.geometry(self.window_size)

    def change_text_size(self):
        #gives options of changing text
        self.clear_window()
        Button(self.root, text="Increase Text", font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color, command=self.increase_text).pack(pady=5)
        Button(self.root, text="Decrease Text", font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color, command=self.decrease_text).pack(pady=5)
        Button(self.root, text="Go Back", font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color, command=self.setup_main_menu).pack(pady=5)

    def increase_text(self):
        #increases the text
        self.clear_window()
        if self.window_size=="1200x900":  # Prevent font size from becoming too large
            if self.font_size_button < 22:
                self.font_size_header += 2
                self.font_size_label += 2
                self.font_size_button += 2
                self.change_text_size()
            else:
                self.clear_window()
                Label(self.root, text="Error! \nText is too Large!", font=("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=10)
                Button(self.root, text="Okay", font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color, command=self.setup_main_menu).pack(pady=5)
        else:
            if self.font_size_button < 14:  # Prevent font size from becoming too large
                self.font_size_header += 2
                self.font_size_label += 2
                self.font_size_button += 2
                self.change_text_size()
            else:
                self.clear_window()
                Label(self.root, text="Error! \nText is too Large!\nChange window size.", font=("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=10)
                Button(self.root, text="Okay", font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color, command=self.setup_main_menu).pack(pady=5)

    def decrease_text(self):
        #decreases the text
        self.clear_window()
        if self.font_size_button > 8:  # Prevent font size from becoming too small
            self.font_size_header -= 2
            self.font_size_label -= 2
            self.font_size_button -= 2
            self.change_text_size()
        else:
            self.clear_window()
            Label(self.root, text="Error! \nText is too Small!", font=("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=10)
            Button(self.root, text="Okay", font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color, command=self.setup_main_menu).pack(pady=5)

        
        
    def ask_stop(self):
        #Makes sure the user wants to exit
        self.clear_window()
        Label(self.root, text="Are you sure you want to exit?", font=("Helvetica", self.font_size_header), bg=self.bg_color, fg=self.text_color).pack(pady=10)
        Button(self.root, text="Yes", font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color, command=self.stop).pack(pady=5)
        Button(self.root, text = "No", font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color, command=self.setup_main_menu).pack(pady=5)

    def stop(self):
        #Tells the user how to exit the game
        self.clear_window()
        Label(self.root, text="Please hit the x on top of the screen", font=("Helvetica", self.font_size_header), bg=self.bg_color, fg=self.text_color).pack(pady=10)

    def login(self):
        #Handle user login
        self.clear_window()

        Label(self.root, text="Enter your username:", font=("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=10)
        self.user_entry = Entry(self.root, font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color)
        self.user_entry.pack(pady=10)

        Label(self.root, text="Enter your password:", font=("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=10)
        self.pass_entry = Entry(self.root, show="*", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color)
        self.pass_entry.pack(pady=10)

        Button(self.root, text="Ok", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.login_user).pack(pady=5)
        Button(self.root, text="Cancel", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command=self.setup_main_menu).pack(pady=5)

    def login_user(self):
        #validate user login
        usern = self.user_entry.get()
        passw = self.pass_entry.get()
        
        self.clear_window()
        
        if usern is None or usern=="" or usern == " ":
                Label(self.root, text="Username can't be blank!", font = ("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=5)
                Button(self.root, text="Try Again", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.login).pack(pady=5)
                Button(self.root, text="Cancel", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.setup_main_menu).pack(pady=5)
                return
        if passw is None or passw=="" or passw==" ":
                Label(self.root, text="Password can't be blank!", font = ("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=5)
                Button(self.root, text="Try Again", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.login).pack(pady=5)
                Button(self.root, text="Cancel", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.setup_main_menu).pack(pady=5)
                return

        cursor.execute("SELECT * FROM Users WHERE username = %s AND password = %s", (usern, passw))
        user = cursor.fetchone()

        if user:
            self.current_user_id = user['user_id']
            Label(self.root, text=f"Logged in Successfully!\nWelcome back {usern}", font = ("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=5)
            Button(self.root, text="Next", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.show_game_or_stats).pack(pady=5)
        else:
            Label(self.root, text="Invalid username or password.", font = ("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=5)
            Button(self.root, text="Try Again", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.login).pack(pady=5)
            Button(self.root, text="Cancel", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.setup_main_menu).pack(pady=5)

    def create_account(self):
        # Handle new account creation
        self.clear_window()  # Clear any previous content in the window

        # Create header label
        Label(self.root, text="Enter a username:", font=("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=10)
        
        self.username_entry = Entry(self.root, font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color)
        self.username_entry.pack(pady=10)

        # Create a label and entry field for the password
        Label(self.root, text="Enter a strong password (10+ characters):", font=("Helvetica", 14), bg=self.bg_color, fg=self.text_color).pack(pady=10)
        self.password_entry = Entry(self.root, show="*", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color)
        self.password_entry.pack(pady=10)

        # Creates buttons to ok the account creation or cancel it
        Button(self.root, text="Ok", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.validate_account_creation).pack(pady=5)
        Button(self.root, text="Cancel", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command=self.setup_main_menu).pack(pady=5)

    def validate_account_creation(self):
        #validates the user's username and password
        username = self.username_entry.get()
        password = self.password_entry.get()

        self.clear_window()
        if username is None or username=="" or username == " ":
            Label(self.root, text=f"Error.\nUsername can't be blank!", font = ("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=5)
            Button(self.root, text="Try Again", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.create_account).pack(pady=5)
            Button(self.root, text="Cancel", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.setup_main_menu).pack(pady=5)
            return
        # Check if username already exists
        cursor.execute("SELECT user_id FROM Users WHERE username = %s", (username,))
        existing_user = cursor.fetchone()
        if existing_user:
            Label(self.root, text=f"Error.\nUsername already exists!", font = ("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=5)
            Button(self.root, text="Try Again", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.create_account).pack(pady=5)
            Button(self.root, text="Cancel", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.setup_main_menu).pack(pady=5)
            return

        if password is None or password=="" or password == " ":
            Label(self.root, text=f"Error.\nPassword can't be blank!", font = ("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=5)
            Button(self.root, text="Try Again", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.create_account).pack(pady=5)
            Button(self.root, text="Cancel", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.setup_main_menu).pack(pady=5)
            return
            
        if len(password) < 10:
            Label(self.root, text=f"Error.\nPassword must be at least 10 characters!", font = ("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=5)
            Button(self.root, text="Try Again", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.create_account).pack(pady=5)
            Button(self.root, text="Cancel", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.setup_main_menu).pack(pady=5)
            return

        try:
            # Insert the new user into the database
            cursor.execute("INSERT INTO Users (username, password) VALUES (%s, %s)", (username, password))
            connection.commit()
            cursor.execute("SELECT user_id FROM Users WHERE username = %s", (username,))
            user = cursor.fetchone()
            self.current_user_id = user['user_id']
            self.clear_window()
            Label(self.root, text=f"Account created Successfully!\nWelcome {username}!", font = ("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=5)
            Button(self.root, text="Next", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.show_game_or_stats).pack(pady=5)
            
        except mysql.connector.Error as err:
            Label(self.root, text=f"Error.\nCould not create account: {err}", font = ("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=5)
            Button(self.root, text="Try Again", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.create_account).pack(pady=5)
            Button(self.root, text="Cancel", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.setup_main_menu).pack(pady=5)

    def play_as_guest(self):
        #Start the game in guest mode
        self.current_user_id = None
        self.show_game_or_stats()

    def leaderboard(self):
        #view the leaderboard
        self.clear_window()
        Label(self.root, text="Leaderboard Options", font=("Helvetica", self.font_size_header, "bold"), bg=self.bg_color, fg=self.text_color).pack(pady=10)
        
        Button(self.root, text="Games Played", font=("Helvetica", self.font_size_button), command = self.show_leaderboard_games_played, bg=self.button_color, fg=self.text_color).pack(pady=10)
        Button(self.root, text="High Score", font=("Helvetica", self.font_size_button), command = self.show_leaderboard_score, bg=self.button_color, fg=self.text_color).pack(pady=10)
        Button(self.root, text = "Back to Main Menu", font=("Helvetica", self.font_size_button), command=self.setup_main_menu, bg=self.button_color, fg=self.text_color).pack(pady=10)
    
    def show_leaderboard_games_played(self):
        self.clear_window()
        Label(self.root, text="Leaderboard - Games Played", font=("Helvetica", self.font_size_header, "bold"), bg=self.bg_color, fg=self.text_color).pack(pady=10)
        
        cursor.execute("""SELECT username, num_times_played FROM Users ORDER BY high_score DESC LIMIT 10;""")
        leaderboard_data=cursor.fetchall()

        for rank, row in enumerate(leaderboard_data, start = 1):
            Label(self.root, text=f"{rank}. {row['username']}-{row['num_times_played']} games", font=("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=2)

        Button(self.root, text = "Back to Leaderboard Options", command=self.leaderboard, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=10)
        Button(self.root, text = "Back to Main Menu", command=self.setup_main_menu, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=10)

    def show_leaderboard_score(self):
        self.clear_window()
        Label(self.root, text="Leaderboard - Highest Score", font=("Helvetica", self.font_size_header, "bold"), bg=self.bg_color, fg=self.text_color).pack(pady=10)
        
        cursor.execute("""SELECT username, high_score FROM Users ORDER BY high_score DESC LIMIT 10;""")
        leaderboard_data=cursor.fetchall()

        for rank, row in enumerate(leaderboard_data, start = 1):
            Label(self.root, text=f"{rank}. {row['username']}-{row['high_score']} points", font=("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=2)

        Button(self.root, text = "Back to Leaderboard Options", command=self.leaderboard, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=10)
        Button(self.root, text = "Back to Main Menu", command=self.setup_main_menu, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=10)
    
    def show_game_or_stats(self):
        #Give option to play the game or view stats after login or guest play
        self.clear_window()
        if self.current_user_id:
            # If a user is logged in, show both options
            Button(self.root, text="Play Game", command=self.start_game, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=5)
            Button(self.root, text="View Stats", command=self.view_stats, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=5)
            Button(self.root, text="Go Back", command=self.setup_main_menu, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=5)
        else:
            # If no user is logged in (guest), only show Play Game option
            Button(self.root, text="Play Game as Guest", command=self.start_game, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=5)
            Button(self.root, text="Go Back", command=self.setup_main_menu, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=5)

    def start_game(self):
        #Initialize the game and display the first question
        self.questions = fetch_questions()
        if not self.questions:
            Label(self.root, text="No questions were found.", font = ("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=5)
            Button(self.root, text="Okay", font=("Helvetica", self.font_size_button), bg=self.bg_color, fg=self.text_color, command = self.setup_main_menu).pack(pady=5)
            return

        self.current_question_index = 0
        self.score = 0
        self.display_question()

    def display_question(self):
        #Display the current question with answer options
        self.clear_window()
        if self.current_question_index >= len(self.questions):
            self.end_game()
            return

        question = self.questions[self.current_question_index]
        
        self.question_var.set(question['question'])
        Label(self.root, text=f"Question {self.current_question_index+1} out of {len(self.questions)}", font=("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=10)
        Label(self.root, textvariable=self.question_var, font=("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color, wraplength=400).pack(pady=10)

        # Display answer buttons
        options = ['a', 'b', 'c', 'd']
        for option in options:
            button = Button(self.root, text=f"{option.upper()}. {question[option]}", 
                            command=lambda opt=option: self.check_answer(opt), font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color)
            button.pack(pady=5)
            self.answer_buttons.append(button)
        Button(self.root, text="Stop Quiz", command=self.end_game, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=20)

    def check_answer(self, selected_option):
        #Check if the selected answer is correct
        self.clear_window()
        question = self.questions[self.current_question_index]
        if selected_option == question['correct_answer']:
            Label(self.root, text = "Correct!\nYour answer was correct.", font=("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=10)
            self.score += 1
        else:
            Label(self.root, text = f"Incorrect. \nThe correct answer was {question['correct_answer']}.", font=("Helvetica", self.font_size_label), bg=self.bg_color, fg=self.text_color).pack(pady=10)

        self.current_question_index += 1
        Button(self.root, text="Next Question", command=self.display_question, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=20)

    def end_game(self):
        #End the game and update the database if necessary
        total_questions = len(self.questions)
        message = f"Your final score is {self.score} out of {total_questions}."
        
        self.clear_window()
        Label(self.root, text=message, font=("Helvetica", self.font_size_header, "bold"), bg=self.bg_color, fg=self.text_color).pack(pady=20)
        
        if self.current_user_id:
            update_user_stats(self.current_user_id, self.score)

        Button(self.root, text="Play Again", command=self.start_game, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=10)
        Button(self.root, text="Main Menu", command = self.setup_main_menu, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=10)                                

    def view_stats(self):
        #View user stats
        self.clear_window()
        cursor.execute("SELECT * FROM Users WHERE user_id = %s", (self.current_user_id,))
        user = cursor.fetchone()

        Label(self.root, 
              text=f"Username: {user['username']}\nHigh Score: {user['high_score']}\nAverage Score: {user['avg_score']:.2f}\nLast Score: {user['last_score']}\nTimes Played: {user['num_times_played']}", 
              font=("Helvetica", self.font_size_label, "bold"), bg=self.bg_color, fg=self.text_color).pack(pady=20)
        Button(self.root, text="Go Back", command=self.show_game_or_stats, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=10)
        Button(self.root, text="Main Menu", command=self.setup_main_menu, font=("Helvetica", self.font_size_button), bg=self.button_color, fg=self.text_color).pack(pady=10)

# Run the application
if __name__ == "__main__":
    root = Tk()
    app = GameApp(root)
    root.mainloop()
```
